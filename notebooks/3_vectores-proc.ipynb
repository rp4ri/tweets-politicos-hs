# Qué hay acá?

Simplemente se toma los vectores que estan en distintos archivos y se los junta en uno solo. Esto es para poder trabajar con todos los vectores de una sola vez.

```python
import pandas as pd
```

```python
politico = 'lfcamacho'
cont = 1

# while there is file to read
while True:
    path = '../data/intermediate/' + politico + '-vectores_' + str(cont) + '.json'
    try:
        # if it is the first file, read it and save it in a dataframe
        if cont == 1:
            df = pd.read_json(path)
        # if it is not the first file, read it and append it to the dataframe
        else:
            df = df.append(pd.read_json(path))
    except:
        break
    cont += 1

# shape
print('Shape: ', df.shape)
df.head()
```

```python
# export to json
df.to_json('../data/processed/' + politico + '-vectores.json', orient='records')
```
